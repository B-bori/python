## **#3**

**- make your small database that only includes the following information:**

**Where both the "released year" was later than 1979 and "Music Recording Sales" was larger than 24.** 

```
filter1 = data["Released"] > 1979
filter2 = data["Music Recording Sales (millions)"]>24

DB = data.where(filter1 & filter2)
DB
```

## **#4**

**- from your own small database, find if their "Genre" includes "Rock"?**  

```
# sol1) 

Genre = list(DB["Genre"])

Genre_later = []

for i in Genre:
  if (type(i)==str):
    Genre_later.append(i.split(', '))

for i in Genre_later:
  if 'rock' in i:
    print("Genre column in DB includes 'rock'")
    break




# sol2)

for i in range(len(DB['Genre'])):
  if(type(DB['Genre'][i])==str):
    if DB['Genre'][i].find('rock') != -1:
      print("Genre column in DB includes 'rock'")
      break




# sol3)

for i in range(len(DB)):
  if(type(DB['Genre'][i])==str):
    if "rock" in DB['Genre'][i]:
      print("Genre column in DB includes 'rock'")
      break
```
